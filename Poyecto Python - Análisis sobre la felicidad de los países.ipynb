```python
# 4.Une ambos dataframes, sin importar que los dos compartan las mismas diferencias.

df = pd.concat([Data15, Data16], ignore_index=True, join='outer')
df.head(5)
```

```python
# 5. Revisa el número de nulos que hay por cada columna, así como su porcentaje.

def nulos(df):
    # Cantidad de nulos
    total = df.isnull().sum()
    # Porcentaje de nulos
    porcentaje = total/len(df)*100
    return total, porcentaje
```

```python
## 5.1 Cantidad de valores nulos por columna.

print("Cantidad de valores nulos por columna")
nulos(df)[0]
```

```python
## 5.2 Porcentaje de valores nulos por columna.

print("Porcentaje de valores nulos por columna")
nulos(df)[1]
```

```python
# 6. Cambia los valores nulos de las columnas "Lower Confidence Interval" y "Upper Confidence Interval" por un número aleatorio entre el valor mínimo y máximo de la misma columna (un único número, no es necesario uno diferente para cada fila con valor nulo).

max = df['Lower Confidence Interval'].max()
min = df['Lower Confidence Interval'].min()
df['Lower Confidence Interval'] = df['Lower Confidence Interval'].fillna(np.random.randint(min, max))
```

```python
max = df['Upper Confidence Interval'].max()
min = df['Upper Confidence Interval'].min()
df['Upper Confidence Interval'] = df['Upper Confidence Interval'].fillna(np.random.randint(min, max))
```

```python
nulos(df)[0]
```

```python
# 7. Cambia los valores nulos de la columna "Standard Error" por su media al cuadrado.

media = df['Standard Error'].mean()**2
df['Standard Error'] = df['Standard Error'].fillna(media)
```

```python
from plotnine import *
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
(
    ggplot(df) +
    geom_point(
        aes(x="Family", y="Health (Life Expectancy)"),
        color = 'red',
        alpha = 0.6,
        size=4
    )
    + labs(title="Relación Familia vs Salud", x="Familia", y="Salud")
    + geom_smooth(
        aes(x="Family", y="Health (Life Expectancy)"),
        color = 'yellow',
        alpha = 0
    )
)
```

```python
# 10. Muestra de forma gráfica la relación entre la puntuación de felicidad y la confianza (corrupción del gobierno).


(
    ggplot(df) +
    geom_point(
        aes(x="Happiness Score", y="Trust (Government Corruption)"),
        color = 'green',
        alpha = 0.6,
        size=4
    )
    + labs(title="Relación Felicidad vs Confianza (corrupción del gobierno)", x="Felicidad", y="Confianza (corrupción del gobierno)")
    + geom_smooth(
        aes(x="Happiness Score", y="Trust (Government Corruption)"),
        color = 'blue',
        alpha = 0.8
    )
)
```

```python
# 12. Tras unir los dataframes, los países aparecerán más de una vez. Muestra agrupado el dataframe por país con el valor máximo de felicidad, sin importar el año.

## 12.1 Ordenar dataframe por país de manera alfabética, para poder obtener una buena visualización.

df = df.sort_values("Country", axis=0)
df.head(20)
```

```python
## 12.2 Nuevo dataframe con la agrupación del país y la puntuación máxima de felicidad.

df_maxima = df.groupby("Country", as_index=False).agg({'Happiness Score' : 'max'})

## 12.3 Filtramos el dataframe con países duplicados por el dataframe con valores máximos.

df = pd.merge(df,df_maxima, on=['Country', 'Happiness Score'], how='inner')
df.head(5)
```

```python
# 13. ¿Tiene relación la felicidad con la generosidad? Muéstralo gráficamente a través de la puntuación de libertad.

(
    ggplot(df) +
    geom_point(
        aes(x="Happiness Score", y="Generosity", colour="Freedom", size='Freedom'),
        alpha = 0.6
    )
    + labs(colour = 'Libertad', size='Libertad')
    + labs(title="Relación Felicidad vs Generosidad vs Libertad", x="Felicidad", y="Generosidad")
    + geom_smooth(
        aes(x="Happiness Score", y="Generosity"),
        color = 'peru',
        alpha = 1
    )
)
```

```python
# 14. Muestra la distribución del grado de distopía en función de la región.


(
    ggplot(df)
    + geom_histogram(
        mapping=aes('Dystopia Residual', fill = 'Region'), color='black'
    )
    + facet_wrap('~Region', ncol = 2)
    + labs(x='Distopía', y='Grado')
    + theme(
        figure_size=(20, 15),
        panel_background=element_rect(fill="pink", alpha=0.06),
        panel_grid=element_line(color="orange", alpha=0.4, size=0.6)
    )
)
```
